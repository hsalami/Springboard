After having cleaned the violation and business datasets, we now explore how the information and attributes of the restaurants are associated with the type of violation detected during inspection of the restaurants.

The inspection and Yelp datasets provide several information about each restaurant, which includes: location, description of the place, starting business date, number of star reviews, total number of reviews, cuisine served, special food served and some characteristics of the place. The inspection dataset also includes inspection time and results (number of low, medium and high violations), and Yelp  dataset includes text reviews for each restaurant. By examining this information, we would like to know which features help in identifying the restaurants that are more likely to make a violation.

In particular, we analyze both datasets and examine if violations are changing across time and season, and if they are following any trends. Do they occur more in restaurants, grocery stores, food facilities in schools, or other places? Are there any differences in the number of violations between chain and non-chain restaurants? In which regions of Allegheny county, the violations are focused? How does the age of restaurants affect their inspection results? Do old restaurants follow regulations more thoroughly? How do popular, high star review and expensive restaurants perform in terms of hygiene? In which cuisine, do violations occur more? Do restaurants with high number of violations serve a specific type of food? Can text reviews help in identifying any possible violation? In this report, we address those questions to gain more insight about the circumstances of each type of violations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
import geopandas as gpd
from shapely.geometry import Point
import geoplot as gplt
import geoplot.crs as gcrs
from pingouin import welch_anova, anova
from scipy.stats import spearmanr
from afinn import Afinn
from dateutil.relativedelta import relativedelta
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
```

We define here the functions that we are going to use in visualizing the features in our data.

```python
# This function plot the mean of violation number vs the maximum number
# of violations detected in each category of the input feature. This is
# done for each violation type (low, medium, high)


def plot_mean_max(df, feature, figsize=(15, 25), legend_visible=True):

    figure, axes = plt.subplots(3, 1, figsize=figsize)

    for viol, ax, color in zip(['low', 'medium', 'high'], axes, ['g', 'b', 'r']):
        viol_max = "{}_max".format(viol)
        viol_mean = "{}_mean".format(viol)
        df.plot.scatter(viol_max, viol_mean, c=color, label=viol, ax=ax)

        # annotate the points on the scatter plot
        for x, y, ft in zip(df[viol_max], df[viol_mean], df[feature]):
            ax.annotate(ft, (x, y), textcoords="offset points", xytext=(0, 10), ha='center',
                        fontsize=11)

        # set axes properties
        ax.set_title('{} Violation'.format(
            viol[0].upper()+viol[1:]), fontsize=14)
        ax.set_xlabel(
            'Maximum Number of Violations per Inspection', fontsize=14)
        ax.legend(loc='upper left', fontsize=12).set_visible(legend_visible)
        ax.set_ylabel(
            'Average Number of Violations per Inspection', fontsize=14)
        ax.set_ylim([ax.get_ylim()[0], ax.get_ylim()[1]+0.1])
        ax.set_ylim([ax.get_ylim()[0], ax.get_ylim()[1]+0.1])

    return figure
```

```python
# This function plots the violin plot of the violation number
# for each category of the feature that is passed as input.
# This is done for each violation type.


def plot_violin(df, feature, figsize=None):

    figure, axes = plt.subplots(1, 3, figsize=figsize)

    for viol, ax in zip(['low', 'medium', 'high'], axes):
        sns.violinplot(data=df, x=feature, y=viol, ax=ax)
        ax.set_xlabel(feature)
        ax.set_ylabel('')
        ax.set_title('{} Violation'.format(viol[0].upper()+viol[1:]))
    axes[0].set_ylabel('Violin Plot of the Number of Violation')
    return figure
```

```python
# This function takes as input a continuous feature, and plot
# the scatter plot of the violation number vs the feature values.
# This is done for each violation type.


def plot_scatter(df, feature, figsize, alpha):

    figure, axes = plt.subplots(1, 3, figsize=figsize)
    viols = ['low', 'medium', 'high']
    colors = ['g', 'b', 'r']

    for ax, viol, c in zip(axes, viols, colors):
        df.plot(feature, viol, marker='o', linestyle='',
                ax=ax, color=c, alpha=alpha)
        ax.set_xlabel(feature, fontsize=14)
        ax.set_ylabel('')
        ax.set_xlim([df[feature].min()-1, df[feature].max()+15])
    axes[0].set_ylabel('Number of Violations', fontsize=14)
    return figure
```

# Types of Violations

We start our analysis by checking the types of violations reported and how they are evolving across time. During each inspection of a given restaurant, all detected violations are classified into low, medium or high violation, and multiple violations can be found during same inspection. Let us check which type of violation is more common and if there is any trend in their occurrences across time.

*How many low, medium and high violations occur per Inspection?*

To check how large each number of violation can be per inspection, we examine their distribution by plotting the box plot of each.

```python
bplt = plt.boxplot(data[['low', 'medium', 'high']].T, vert=True, labels=[
                   'low', 'medium', 'high'], patch_artist=True)
colors = ['green', 'blue', 'red']
for patch, color in zip(bplt['boxes'], colors):
    patch.set_facecolor(color)
plt.ylabel("Number of Violations per Inspection")
plt.xlabel("Type of Violation")
```

We see that the number of low violation per inspection is more spread out than that of medium and high violations. Also, the number of low violations found in an inspection can be larger than that of medium and high violations. We also notice that more than 50% of the number of high violations is zero, which means that they are less common than the other two types of violations. We can further check how common each type of violation is, by computing their non-zero percentage.  

```python
print("Percentage of non-zero low violations:",
      100*len(data[data.low > 0])/len(data))
print("Percentage of non-zero medium violations:",
      100*len(data[data.medium > 0])/len(data))
print("Percentage of non-zero high violations:",
      100*len(data[data.high > 0])/len(data))
```

*How are violations changing over time?*

Let us now check the evolution of the number of each violation type with time, to see if there is an increase or decrease in their occurrence.

We count the average number of each type of violation for each month across the years.

```python
# Count the number of violations for each month
data['inspect_dt'] = pd.to_datetime(data['inspect_dt'])
hist = data[['inspect_dt', 'low', 'medium', 'high']]
hist = hist.assign(inspect_my=hist['inspect_dt'].apply(
    lambda x: pd.to_datetime(str(datetime.date(x.year, x.month, 1)))))
trend = hist.groupby(["inspect_my"])[['low', 'medium', 'high']].mean()
```

We then plot the evolution of the average number of violations.

```python
trend.plot(figsize=(10, 5), color=[
           'green', 'blue', 'red'], marker='o', linestyle='')
plt.ylabel('Average Number of violations per Inspection', fontsize=12)
plt.xlabel('Inspection Month')
plt.title('Evolution of Average Number of each Violation Type')
```

We see that on average the number of each type of violations is not decreasing with time, which implies the importance of doing regular inspections for the food facilities and restaurants. 

We extract the details of the inspection date and time, by adding the columns month, day, day name, year and hour. 

```python
data = data.assign(month=data.inspect_dt.apply(lambda s:s.month))
data = data.assign(day=data.inspect_dt.apply(lambda s:s.day))
data = data.assign(day_name=data.inspect_dt.apply(lambda s:s.day_name()))
data = data.assign(year=data.inspect_dt.apply(lambda s:s.year))
```

```python
data = data.assign(hour=pd.to_datetime(data.start_time).apply(lambda s:s.hour))
```

# Place Description for each Violation Type

The column 'description' provides a brief description of the facility inspected. More specifically, it specifies whether the inspected place is a restaurant, supermarket or convenience store, and whether it belongs to a chain and if it provides liquor. We now check in which places each type of violation occur more.

For each place category indicated by the description column, we compute the average and maximum number of each violation type detected per inspection.

```python
desc_mean = data.groupby(["description"])[
    ['low', 'medium', 'high']].mean().reset_index()
desc_max = data.groupby(["description"])[
    ['low', 'medium', 'high']].max().reset_index()
description = desc_mean.merge(
    desc_max, on='description', suffixes=['_mean', '_max'])
```

We now visualize the obtained averages and maximum values for each place type. We first shorten some of the names of places (of column 'description'), because we will use them to annotate the points on the plots.

```python
description['description'] = description.description.apply(
    lambda s: s.replace('Restaurant', "Rest."))
description['description'] = description.description.apply(
    lambda s: s.replace('Retail/Convenience', 'Retail'))
description['description'] = description.description.apply(lambda s: s.replace(
    'Nursing Home/Personal Care Comb.', 'Nursing Home'))
description['description'] = description.description.apply(lambda s: s.replace(
    'Kitchen', 'Kit.'))
```

To make the visualization clearer to read, we provide a separate plot for each violation type. In each plot, we plot the average number of violations vs the maximum number of violations for each place category.

```python
figure = plot_mean_max(description, 'description')
```

In the above plots, the less clean food facilities appear in the upper right corner of the plot and the cleaner facilities appear in the lower left corner of the plot. We notice that for all violation types, the high numbers of violations are mainly detected in restaurants. On the other hand, cleaner places are those that provide child food or packaged foods or those that are inside a school or a warehouse. Moreover, we notice that "rooming house" and mobile restaurants are places with large means of medium violations.

## Violation in Chain/Non-Chain Places

We now zoom into the chain and non chain places, to check if this characteristic can be associated with the type of violation.

We first extract the places that belong to a chain and those that do not belong. We then add an indicator column that specifies this characteristic. 

```python
# list of categories of non-chain places
non_chain = ['Bakery', 'Packaged Food Only', 'Restaurant with Liquor',
             'Restaurant without Liquor',
             'Retail/Convenience Store', 'Supermarket']
# list of categories of chain places
chain = ['Chain Bakery', 'Chain Packaged Food Only', 'Chain Restaurant with Liquor',
         'Chain Restaurant without Liquor', 'Chain Retail/Convenience Store',
         'Chain Supermarket']

# extract the description and the the type of violation
chain_data = data[['description', 'low', 'medium', 'high']]

# find the places that belong to a chain
chains = chain_data['description'].apply(
    lambda s: 'True' if s in chain else 'False')
# assign a new column that indicates if the palces corresponds to a chain or not
chain_data = chain_data.assign(chain=chains)
chain_data = chain_data[((chain_data.description.isin(chain)) |
                         (chain_data.description.isin(non_chain)))]
```

We then plot the violin plots for each type of violation and then compare them between the chain and non-chain places.

```python
chain_data_melted = chain_data.melt(['description', 'chain'],
                                    var_name='violation', value_name='number')
sns.violinplot(data=chain_data_melted, x='violation', y='number', hue='chain')
plt.xlabel('Type of Violation')
plt.ylabel('Violin Plot of Each Violation Type')
plt.show()
```

We notice that the median of medium violations is slightly higher for non-chain places, and that for chain places the distribution of low violations is more right skewed. This might be because chain restaurants have to follow strict corporate standards.

# Historical Information of Past Inspections

In order to enrich more our data, we create additional columns that capture past historical performance of each inspected place and that include more information about the inspection instance. By adding these columns, we try to examine if there is any trend in the evolution of the number of violations detected. Do restaurants do better than the last time they were inspected? Do regular inspections help in reducing the number of violations? 
We next list the added columns.

1. *Inspection number*: it specifies whether it is the first inspection for the restaurant, or the second and so on.  

```python
# Add the inspection number as a column
data = data.sort_values('inspect_dt')
grouped = data[['id','encounter','inspect_dt']].groupby([
    'id','encounter','inspect_dt']).count().reset_index(['encounter','inspect_dt'])
counts = data.groupby('id')['encounter'].count()

nums = [] 
for c in list(counts):
    nums = nums + list(range(1,c+1))

grouped = grouped.assign(inspect_num=nums)

data = data.merge(grouped[['encounter','inspect_num']])
```

2. *Number of days between inspections*: the number of days passed since last inspection day.

```python
# How much time between each inspection?
data = data.assign(days_between =(
    data.sort_values('inspect_dt').groupby('id')[['inspect_dt']].diff()).apply(
    lambda s:s.dt.days).fillna(0))
```

3. *Statistics of Past performance*: columns corresponding to lagged values and to rolling statistics of past low, medium and high violations. These columns include the max, median and standard deviations of all past violations detected before the present inspection. They also include the last numbers of low, medium and high violation detected in the last three previous inspection instances and their log ratios as well.

```python
# Add historical stat info
data = data.assign(last_low=np.nan)
data = data.assign(last_medium=np.nan)
data = data.assign(last_high=np.nan)

data = data.assign(last2_low=np.nan)
data = data.assign(last2_medium=np.nan)
data = data.assign(last2_high=np.nan)

data = data.assign(last3_low=np.nan)
data = data.assign(last3_medium=np.nan)
data = data.assign(last3_high=np.nan)

data = data.assign(median_low=np.nan)
data = data.assign(median_medium=np.nan)
data = data.assign(median_high=np.nan)

data = data.assign(max_low=np.nan)
data = data.assign(max_medium=np.nan)
data = data.assign(max_high=np.nan)

data = data.assign(std_low=np.nan)
data = data.assign(std_medium=np.nan)
data = data.assign(std_high=np.nan)
```

```python
ids = data.id.unique()
for id_place in ids:
    dates = data[data.id==id_place].inspect_dt
    for dt_prev, dt_now in zip(dates[:-1],dates[1:]):
        prev = np.logical_and((data.id==id_place), (data.inspect_dt==dt_prev))
        to_look = np.logical_and((data.id==id_place), (data.inspect_dt<dt_now))
        now = np.logical_and((data.id==id_place), (data.inspect_dt==dt_now))
        data.at[now,'last_low'] = int(data[prev]['low'])
        data.at[now,'last_medium'] = int(data[prev]['medium'])
        data.at[now,'last_high'] = int(data[prev]['high'])

        data.at[now,'median_low'] = data[to_look]['low'].median()
        data.at[now,'median_medium'] = data[to_look]['medium'].median()
        data.at[now,'median_high'] = data[to_look]['high'].median()
        
        data.at[now,'std_low'] = data[to_look]['low'].std()
        data.at[now,'std_medium'] = data[to_look]['medium'].std()
        data.at[now,'std_high'] = data[to_look]['high'].std()
        
        data.at[now,'max_low'] = data[to_look]['low'].max()
        data.at[now,'max_medium'] = data[to_look]['medium'].max()
        data.at[now,'max_high'] = data[to_look]['high'].max()
    for dt_prev, dt_now in zip(dates[:-2],dates[2:]):
        prev = np.logical_and((data.id==id_place), (data.inspect_dt==dt_prev))
        now = np.logical_and((data.id==id_place), (data.inspect_dt==dt_now))
        data.at[now,'last2_low'] = int(data[prev]['low'])
        data.at[now,'last2_medium'] = int(data[prev]['medium'])
        data.at[now,'last2_high'] = int(data[prev]['high'])
    for dt_prev, dt_now in zip(dates[:-3],dates[3:]):
        prev = np.logical_and((data.id==id_place), (data.inspect_dt==dt_prev))
        now = np.logical_and((data.id==id_place), (data.inspect_dt==dt_now))
        data.at[now,'last3_low'] = int(data[prev]['low'])
        data.at[now,'last3_medium'] = int(data[prev]['medium'])
        data.at[now,'last3_high'] = int(data[prev]['high'])
```

We also add the ratio of past number of violations (low, medium and high).

```python
data = data.assign(ratio_last_low=np.log((1+data.last_low)/(1+data.last2_low)))
data = data.assign(ratio_last_medium=np.log(
    (1+data.last_medium)/(1+data.last2_medium)))
data = data.assign(ratio_last_high=np.log(
    (1+data.last_high)/(1+data.last2_high)))
```

We end up with the following additional numerical columns. We now compute the Spearman's correlation between each of these numerical columns and the number of low, medium and high violations.

```python
cols= ['days_between', 'last_low', 'last_medium', 'last_high',
       'last2_low', 'last2_medium', 'last2_high', 'last3_low', 'last3_medium',
       'last3_high', 'median_low', 'median_medium', 'median_high', 'max_low',
       'max_medium', 'max_high', 'std_low', 'std_medium', 'std_high', 'ratio_last_low',
       'ratio_last_medium', 'ratio_last_high','age']
```

```python
results=[]
for col in cols:
    result=[]
    for viol in ['low', 'medium', 'high']:
        res = spearmanr(data[viol], data[col],nan_policy='omit')
        result.append(res.correlation)
    results.append(result)
        
fig, ax = plt.subplots(figsize=(8,7))
sns.heatmap(pd.DataFrame(results, columns=['low','medium','high'],index=cols),yticklabels=True,
            annot=True,ax=ax)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.title("Spearman's Correlation")
#fig.savefig("corr.png",bbox_inches='tight')
plt.show()
```

We observe the following:
- For low violation, the strongest positive correlation is shown to be with its own past statistics: max_low and std_low, as well as with the number of days passed since last inspection. The strongest negative correlation is shown to be with the log ratio of past medium and high violations;
- For Medium violations, the strongest positive correlation is shown to be with the number of days passed since last inspection. The strongest negative correlation is shown to be with the log ratio of past high violations;
- For high violations, the strongest positive correlation is shown to be with its own statistics: max_high, std_high. The strongest negative correlation is shown to be with the log ratio of past low and medium violations.

Each type of violation shows to have positive correlation with its past performance translated by the rolling maximum and standard deviation measures. Each type of violation also shows negative correlation with the log ratio of past violations of the other type. Also the number of days between two consecutive inspections positively correlates with each type of violation, this might imply the importance of regular inspections.

# Topic Modeling

We now perform topic modeling on the text reviews to check the main topics mentioned in the reviews and represent the set of reviews mentioned before each inspection with a representative vector of features.

For each inspection, we join the texts of all reviews that were mentioned for the inspected place before the inspection date.

```python
# merge the inspection data with the reviews
grouped = subset.merge(reviews, on='business_id')

to_choose = (grouped.date < pd.to_datetime(grouped.inspect_dt))
grouped = grouped[to_choose]
greviews= grouped.groupby('encounter')['clean_text'].apply(' '.join).reset_index()
```

```python
greviews = greviews.merge(data[['encounter','low','medium','high']])
```

To find the topics, we can either use LDA (Latent Dirichlet Allocation) or NMF (Non-Negative Matrix Factorization). We are going to go with second method (NMF) using the NMF module form sklearn library. 

NMF algorithm takes as input the number of topics and documents (set of reviews in our case) and returns the words that constitute each topic and specifies to which topic each review belongs. This is done by first finding a matrix representation of all reviews, where each row represents one review and each column contains the frequency of each word in the whole set of reviews. This matrix is then decomposed into two smaller matrices, where the first one maps each review to one topic and the second specifies the weight of each word for each topic. 

We first represent, for each inspection, the set of joined reviews with a vector of size 500 using TfidfVectorizer; this vector represents the frequency of the words mentioned in the reviews of each inspected place.

```python
no_features = 500
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, max_features=no_features)

tfidf = tfidf_vectorizer.fit_transform(greviews.clean_text)
tfidf_feature_names = tfidf_vectorizer.get_feature_names()
```

```python
tfidf.shape
```

Now each inspection instance has a vector of size 500 representing the text reviews. To extract the main topics, we decompose the obtained matrix using NMF in  order to reduce the word feature space from 500 to 5 (i.e., 5 topics). 

```python
no_topics = 5

nmf = NMF(n_components=no_topics, random_state=1,
              alpha=.1, l1_ratio=1, init='nndsvd').fit(tfidf)
```

We now display the top 10 words in each topic for each type of violation.

```python
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic %d:" % (topic_idx+1))
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))

no_top_words = 10
display_topics(nmf, tfidf_feature_names, no_top_words)
```

Now, for each inspection, we use the decomposition we found using NMF to replace each word vectors of size 500 to a vector of dimension 5, where each column corresponds to the above topic.

```python
res = nmf.transform(X=tfidf)
```

```python
tpcs = pd.concat([greviews,pd.DataFrame(res,columns=['V1','V2','V3','V4','V5'])],axis=1)
```

```python
data= data.merge(tpcs[['encounter','V1','V2','V3','V4','V5']], how='left')
```

Let us check the how the vector of topics of each inspection correlate with low, medium and high violations.

```python
results=[]
cols=['V1','V2','V3','V4','V5']
for col in cols:
    result=[]
    for viol in ['low', 'medium', 'high']:
        res = spearmanr(data[viol], data[col],nan_policy='omit')
        result.append(res.correlation)
    results.append(result)
        
fig, ax = plt.subplots(figsize=(7,7))
sns.heatmap(pd.DataFrame(results, columns=['low','medium','high'],index=cols),yticklabels=True, ax=ax,
           annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.title("Spearman's Correlation")
#fig.savefig('tpc.png')
plt.show()
```

Topic V4 shows the lowest negative correlation with each type of violation. The main words that constitute topic V4 are: coffee, shop and store. This might mean that the inspected place corresponds to a coffee shop or it is a place for shopping, more than it is a restaurant. Low violation has its most significant positive correlation with topic V3, which seems to be related to the service provided by the place. Topic V5 which is consistitued of the words: chinese and suhi roll has greater correlation with each type of violation than topics V1 and V2.
